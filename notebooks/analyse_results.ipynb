```python
import pandas as pd
from collections import Counter, defaultdict
```

```python
df = pd.read_json("../data/initial_runs/initial_runs_combined.jsonl", lines=True)
df
```

```python
c = Counter()

general = df['general'].to_list()
general = [[word.lower() for word in doc] for doc in general]
for doc in general:
    c.update(doc)
         
len(c)
```

```python
# Create a list of keys to delete
keys_to_delete = [key for key, count in c.items() if count < 6]

# Delete the keys from the Counter
for key in keys_to_delete:
    del c[key]

len(c)
```

```python
new_dict = defaultdict(int)

for key, count in c.items():
    key = key.replace(" tone", "")
    key = key.replace(" style", "")
    key = key.replace(" content", "")
    key = key.replace(" language", "")
    key = key.strip()
    new_dict[key] += count
    
len(new_dict)
```

```python
# Sort the items in the defaultdict by their values in descending order
sorted_items = sorted(new_dict.items(), key=lambda item: item[1], reverse=True)

# Print the items with the highest values
for key, value in sorted_items[:200]:  # Adjust the number to print more or fewer items
    print(f"{key}: {value}")
```

```python
total = 0

for key, value in new_dict.items():
    if "QA" in key or "instructions" in key:
        total += value
        print(f"{key}: {value}")
        
print(total)
```

```python
import pandas as pd
import json

# Load the JSONL file into a pandas DataFrame
file_path = '/home/otto-tarkka/Documents/work/LLM_document_descriptors/data/initial_runs/initial_runs_combined.jsonl'
data = []
with open(file_path, 'r') as file:
    for line in file:
        data.append(json.loads(line))

df = pd.DataFrame(data)

# Ensure that the "general" column does not contain empty lists
df['general'] = df['general'].apply(lambda x: x if isinstance(x, list) and len(x) > 0 else ['empty'])

# Extract the "general" column and apply one-hot encoding
general_df = pd.DataFrame(df['general'].tolist()).stack().str.get_dummies().groupby(level=0).sum()

general_df
```

```python
# Calculate the sum of each row in the general_df DataFrame
general_df.sum(axis=1).value_counts()
```

```python
unique_descriptors = set()
for descriptors in df['general']:
    unique_descriptors.update(descriptors)

unique_descriptors = list(unique_descriptors)
print(len(unique_descriptors))
print(unique_descriptors)
```

```python
with open('../data/initial_runs/all_unique_descriptors.txt', 'w') as file:
    for desc in unique_descriptors:
        file.write(f"{desc}\n")
```
