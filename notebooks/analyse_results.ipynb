```python
import pandas as pd
import json

# Load the JSONL file into a pandas DataFrame
file_path = '/home/otto-tarkka/Documents/work/LLM_document_descriptors/data/initial_runs/initial_runs_combined.jsonl'
data = []
with open(file_path, 'r') as file:
    for line in file:
        data.append(json.loads(line))

df = pd.DataFrame(data)

# Ensure that the "general" column does not contain empty lists
df['general'] = df['general'].apply(lambda x: x if isinstance(x, list) and len(x) > 0 else ['empty'])

# Extract the "general" column and apply one-hot encoding
general_df = pd.DataFrame(df['general'].tolist()).stack().str.get_dummies().groupby(level=0).sum()

general_df
```

```python
# Calculate the sum of each row in the general_df DataFrame
general_df.sum(axis=1).value_counts()
```

```python
# Apply MCA to the general_df DataFrame
from prince import MCA

mca = MCA(
    n_components=100,
    n_iter=3,
    copy=True,
    check_input=True,
    engine='sklearn',
    random_state=42,
    one_hot=False # Set to True if the input data is not already one-hot encoded
)

mca = mca.fit(general_df)

mca.eigenvalues_summary
```
