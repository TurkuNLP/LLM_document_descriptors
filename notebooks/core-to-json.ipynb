```python
from pathlib import Path
import json
```

```python
core_path = Path("../data/core/train.tsv")

data = []
with core_path.open("r") as f:
    for line in f:
        fields = line.split("\t")
        assert len(fields) == 3, "Num fields is off."
        entry = {
            "label": fields[0],
            "id": fields[1],
            "text": fields[2],
        }
        data.append(entry)
```

```python
save_path = Path("../data/core/train.jsonl")
with save_path.open("w") as f:
    for line in data:
        f.write(json.dumps(line, ensure_ascii=False) + "\n")
```
