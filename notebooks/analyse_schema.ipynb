```python
import json
from pathlib import Path
import random
```

```python
schema_path = Path("../results/final_schema/descriptor_schema.jsonl")
traces_path = Path("../results/final_schema/descriptor_schema_traces.jsonl")
og_descriptors = Path("../results/new_descriptors/grouped/descriptors_with_ids.jsonl")

def read_jsonl(path):
    out = []
    with path.open("r") as f:
        for line in f:
            obj = json.loads(line)
            out.append(obj)

    return out

schema = read_jsonl(schema_path)
traces = read_jsonl(traces_path)
og = read_jsonl(og_descriptors)
```

```python
root_counts = [l["root_count"] for l in traces]
weights = [l["weight"] for l in traces]
```

```python
sum(1 for i in weights if i == 1)
```

```python
sum(1 for i in root_counts if i == 1)
```

```python
for idx, line in enumerate(traces):
    print(f"{line['descriptor']} - {line['explainer']} - {line['weight']}")
    print()
    print("Root descriptors for this final:")
    sample_of_root_ids = random.sample(line["root_ids"], min(10, len(line["root_ids"])))
    for j, root_id in enumerate(sample_of_root_ids):
        for desc in og:
            if root_id == desc["id"]:
                print(f"    {desc['descriptor']} - {desc['explainer']}")
                break
        if j == 9:
            break
    print("===================================================")
    print()
    if idx == 9:
        break
```
